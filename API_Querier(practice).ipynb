```python
import requests
```

```python
def retrieval_api(api_url):
  response = requests.get(api_url)
  # print('Status Code: ' + str(response.status_code))
  
  # return str(response.json())
  return response.json()
```

```python
test_url= 'https://opendata.cwb.gov.tw/api/v1/rest/datastore/F-C0032-001?Authorization=CWB-B005D345-A59F-4E55-8F01-0EB971BB5BC1&format=JSON&locationName=%E6%96%B0%E5%8C%97%E5%B8%82&sort=time'
```

```python
test_json =  retrieval_api(test_url)
print(type(test_json))

title = test_json['records']['datasetDescription']
location = test_json['records']['location'][0]['locationName']
list = []
# print(test_json['records']['location'][0]['weatherElement'][0]['time'][2])#['parameter']['parameterName'])
for i in range(len(test_json['records']['location'][0]['weatherElement'][0]['time'])):
  list.append(test_json['records']['location'][0]['weatherElement'][0]['time'])
print(title)
print(location)
for i in range(len(list)):
  print(list[0][i]['startTime']+' ~ '+list[0][i]['endTime']+'\t'+list[0][i]['parameter']['parameterName'])
```
